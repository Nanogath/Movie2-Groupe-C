*On peut identifier chaque film de **"movies"** avec l'attribut **"movieId"**. On y applique donc une **clé primaire**.* 

*Un film de **"movies"**, identifié par l'attribut **"movieId"**, est forcément lié à **une seule valeur "imdb"** et **"tmdb"** de la table **"links"**.*

*Comme on retrouve l'attribut **"movieId"** dans la table **"links"**. On y applique une **clé primaire et étrangère**.*

*Les tuples de la table **"ratings"** ne sont pas identifiables avec le seul attribut **"userId"** car un même utilisateur a pu réalisé plusieurs évaluations. 

*De la même façon, l'attribut **"movieId"** ne suffit pas en identifiant unique car différents utilisateurs ont pu évaluer le même film.*

*Bien que peu probable, on peut aussi considérer que plusieurs évaluations aient été ajoutées au même instant, au même **"timestamp"** donc, le rendant inutilisable en clé primaire.*

*Enfin l'attribut **"rating"** est évidemment lui aussi non unique. Notre **clé primaire** sera donc composée de deux attributs: **"userId"** et **"movieId"**, car un utilisateur unique a pu évaluer un film unique une seule fois.*

*Si chaque tuple de la table **"ratings"** est forcément lié à un tuple de la table **"movies"**, on peut considérer la **multiplicité** dans le sens inverse, comme un film possèdant de 0 à une infinité d'évaluations, selon le nombre d'utilisateur ayant évalué un même film.*

*Cette même contrainte est transférable sur la relation **"movies"** et **"tags"**. En l'occurence, un film peut posséder de 0 à une infinité de tags selon le nombre d'utilisateurs l'ayant commenté. En revanche, un tag est propre à un seul film.*

*Afin d'identifier ce tag, sachant que plusieurs films peuvent avoir le même tag, nous appliquerons une **clé primaire** aux attributs **"userId"** et **"movieId"**.*

*Enfin, nous avons placé toutes les **clés étrangères** sur  l'attribut **"movieId"** de chaque table pour les lier par un identifiant commun.*
